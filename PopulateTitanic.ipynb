```python
# import necessary libraries 

import win32com.client
import os
import pandas as pd
import numpy as np
```

```python
# read from titanic.csv

sourceFile = r'C:\Users\ewerfel\Documents\Collabsphere\titanic.csv'

titanicDataFrame = pd.read_csv(sourceFile, header=0, keep_default_na=False)

# Age includes some empty values, which forces the column to string.  This will change values to numeric. 
titanicDataFrame["Age"] = pd.to_numeric(titanicDataFrame["Age"], errors="coerce")

# This will change values that couldn't convert to numeric to the empty string, rather than NaN, which Domino can't use
titanicDataFrame = titanicDataFrame.fillna('')

  
```

```python
notesSession = win32com.client.Dispatch(r'Notes.NOTESSESSION')
collabsphereDb = notesSession.GetDatabase("Dev-01", "developers\ewerfel\Collabsphere\collabsphere3.nsf")

print(collabsphereDb.Title)

```

```python
titanicDataFrame.describe()
```

```python
formName = "Passenger"

#Populate Notes with the first 10 rows in the dataframe

for index, row in titanicDataFrame.T.iloc[:,:10].items():  
    doc = collabsphereDb.CreateDocument
    doc.ReplaceItemValue("Form",formName)
    for i in range(len(row)):
        doc.ReplaceItemValue(row.index[i], row[i])
    doc.Save(1,1)

 
```

```python
#Populate the database with the rest of the rows
for index, row in titanicDataFrame.T.iloc[:,10:].items():  
    doc = collabsphereDb.CreateDocument
    doc.ReplaceItemValue("Form",formName)
    for i in range(len(row)):
        doc.ReplaceItemValue(row.index[i], row[i])
    doc.Save(1,1)   
```
